```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("data/bigTable.csv")

fig = px.choropleth(data, locations='Province_State', locationmode="USA-states", color='Confirmed',
                           color_continuous_scale="Viridis",
                           range_color=(0, max(data["Confirmed"])),
                           scope="usa",
                           labels={'Confirmed':'Confirmed Deaths'})

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
